```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('IMG_98.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
# Apply Median Blur to mask text and smoothen edges
med = cv2.medianBlur(gray, 11)
plt.imshow(med, cmap='gray')
```

```python
# Apply Gaussian Blur for Canny Filter
blur = cv2.GaussianBlur(med, (11,11), 2)
plt.imshow(blur, cmap='gray')
```

```python
# Canny Edge Detection
canny = cv2.Canny(med, 30, 100, 3)
plt.imshow(canny, cmap='gray')
```

```python
# Dilation then Erosion applied for edge enhancement and refinement
dilated = cv2.dilate(canny, (2, 2), iterations=6)
erosion = cv2.erode(dilated, (3, 3), iterations=2)
plt.imshow(erosion, cmap='gray')
```

```python
# Counting contours of eroded image
(cnt, hierarchy) = cv2.findContours(
    erosion.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 1)

plt.imshow(rgb)
```
